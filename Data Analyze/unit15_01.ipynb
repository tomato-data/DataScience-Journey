```python
import pandas as pd

df = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area', header = 0, index_col = 0)
no_notes = df[1].iloc[:, :5]
no_notes.columns = ['나라', '전체 면적', '땅 면적', '물 면적', '물 비율(%)']
# ascending을 True로 하면 오름차순 정렬
print(no_notes.sort_values('물 비율(%)', ascending=False))
no_notes.to_excel('나라별 물비율 내림차순 정렬.xlsx')
```
