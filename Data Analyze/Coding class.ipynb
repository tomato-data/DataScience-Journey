```python
import csv
import matplotlib.pyplot as plt

f = open('Japan Census.csv')
data = csv.reader(f)
m = []
f = []
xtick_name = ['0~4', '5~9', '10~14', '15~19', '20~24', '25~29', '30~34', '35~39', '40~44', '45~49', '50~54', \
          '55~59', '60~64', '65~69', '70~74', '75~79', '80~84', '85~89', '90~94', '95~99', '100+']

for row in data:
    if row[0] == '2021000606' and row[1] == '総人口':
        m.append(float(row[7]))
        f.append(float(row[8]))

m = m[1:22]
f = f[1:22]

plt.plot(xtick_name, m, label='male')
plt.plot(f, label='female')
# 가로로 쓰면 잘 안보이니까 세로로 바꾸자
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```

```python
result = []
for i in range(len(m)):
    result.append(m[i] - f[i])
    
plt.bar(xtick_name, result)
plt.xticks(rotation='vertical')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('Japan Census.csv')
data = csv.reader(f)
m = []
f = []
xtick_name = ['0~4', '5~9', '10~14', '15~19', '20~24', '25~29', '30~34', '35~39', '40~44', '45~49', '50~54', \
          '55~59', '60~64', '65~69', '70~74', '75~79', '80~84', '85~89', '90~94', '95~99', '100+']

for row in data:
    if row[0] == '2021000606' and row[1] == '総人口':
        m.append(-float(row[7]))
        f.append(float(row[8]))

m = m[1:22]
f = f[1:22]

plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('Japan Population Pyramid')
plt.style.use('ggplot')
plt.barh(xtick_name, m, label='male')
plt.barh(xtick_name, f, label='female')
plt.legend()
plt.show()
```
