년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
def change_date(x):
    import datetime
    date = x.split(':')[0]
    hour = x.split(':')[1]
    if hour == '24':
        hour = '00:00:00'
        final_date = pd.to_datetime(date+' '+hour) + datetime.timedelta(days=1)
    else:
        hour = hour + ':00:00'
        final_date = pd.to_datetime(date+' '+hour)
    return final_date

df['(년-월-일:시)'] = df['(년-월-일:시)'].map(change_date)
df['(년-월-일:시)'].head(10)
```
