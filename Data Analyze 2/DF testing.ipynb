```python
import pandas as pd

ns_df = pd.read_csv('ns_book7.csv', low_memory=False)
ns_df.head()
print(len(ns_df))
```

```python
sum(ns_df.duplicated(subset=['도서명', '저자', 'ISBN', '권']))
```

```python
count_df = ns_df[['도서명', '저자', 'ISBN', '권', '대출건수']]
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False)
group_df.head()
```

```python
loan_count = group_df.sum()
loan_count.head()
print(len(loan_count))
```

```python
dup_rows = ns_df.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
unique_rows = ~dup_rows
ns_book3 = ns_df[unique_rows].copy()
```

```python
sum(ns_book3.duplicated(subset=['도서명', '저자', 'ISBN', '권']))
```

```python
ns_book3.head()
```

```python
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_book3.head()
```

```python
ns_book3.update(loan_count)
ns_book3.head()
```

```python
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
ns_book4 = ns_book4[ns_df.columns]
ns_book4.head()
```

```python
# 누락된 값을 찾아 불리언 배열로 반환
set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
# 누락된 값을 빈 문자열로 바꿈
ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''
# 누락된 값이 몇 개인지 셉니다
ns_book4['세트 ISBN'].isna().sum()
```
