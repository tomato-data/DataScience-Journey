```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5) # -5 ~ 5 를 50 부분으로 나눔

print(x)
print(len(x))
```

```python
y_1 = 2 * x

plt.plot(x, y_1)
plt.show()
```

```python
y_1 = 2 * x
y_2 = 3 * x

# label for the axis
plt.xlabel("x value", size = 14)
plt.ylabel("y value", size = 14)

# title of the graph
plt.title("My Graph")

# visible grid
plt.grid()

# show legned and set different style for lines when plotting
plt.plot(x, y_1, label = "y1")
plt.plot(x, y_2, label = "y2", linestyle = "dashed")
plt.legend()

plt.show()
```

```python
x = np.linspace(-10, 10)
y_1 = (x - 2) ** 2 + 3
y_2 = (x + 3) ** 2 - 10

# label for the axis
plt.xlabel("x value", size = 14)
plt.ylabel("y value", size = 14)

# title of the graph
plt.title("My Graph")

# visible grid
plt.grid()

# show legned and set different style for lines when plotting
plt.plot(x, y_1, label = "y1")
plt.plot(x, y_2, label = "y2", linestyle = "dashed")
plt.legend()

plt.show()
```
