```python
eta = 0.1

def polynomial(x, params):
    poly = 0
    for i in range(len(params)):
        poly += params[i]*x**i # 식 1
    return poly

# len(params) = 2
def logarithm(x, params):
    log = 0
    for i in range(len(params)):
        log += params[i]*np.log(x)**i

# --- 각 파라미터의 기울기 ---
def grad_params(X, T, params):
    grad_ps = np.zeros(len(params))
    for i in range(len(params)):
        for j in range(len(X)):
            grad_ps[i] += (polynomial(X[j], params) - T[j])*X[j]**i # 식 7
    return grad_ps

def grad_params_log(X, T, params):
    grad_ps = np.zeros(len(params))
    for i in range(len(params)):
        for j in range(len(X)):
            grad_ps[i] += (logarithm(X[j], params) - T[j])*X[j]**i # 식 7
    return grad_ps

# --- 학습 ---
def fit(X, T, degree, epoch): # degree: 다항식의 차수 epoch: 반복하는 회수
    
    # --- 파라미터 초깃값을 설정 ---    
    params = np.random.randn(degree+1) # 파라미터의 초깃값
    for i in range(len(params)):
        params[i] *= 2**i
        # 고차의 항일수록 파라미터를 크게
        
    # --- 파라미터의 갱신 ---
    for i in range(epoch):
        params -= eta * grad_params(X, T, params) # 식 4
        
    return params

def fit_log(X, T, epoch): # degree: 다항식의 차수 epoch: 반복하는 회수
    
    # --- 파라미터 초깃값을 설정 ---    
    params = np.random.randn(2) # 파라미터의 초깃값
    for i in range(len(params)):
        params[i] *= 10000
        # 고차의 항일수록 파라미터를 크게
        
    # --- 파라미터의 갱신 ---
    for i in range(epoch):
        params -= eta * grad_params_log(X, T, params) # 식 4
        
    return params

# --- 결과표시 ---
# O(n^2)
txt = "Degree: " + str(2)
params = fit(X, T_S, 2, 1)
Y = polynomial(X, params) # 학습 후의 파라미터를 사용한 다항식
plt.title(txt)
plt.scatter(X, T_S)
plt.plot(X, Y, linestyle="dashed")

plt.show()
```
