```python
import numpy as np
import matplotlib.pyplot as plt

a = 1.5 # constant
x = np.linspace(-1, 1)
y = a * x

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return 3 * x + 2

x = 4 # global, not the factor announced in the function
y = my_func(x)
print(y)
```

```python
def func2(x):
    a = 3
    return x ** a

x = np.linspace(0, 2)
y = func2(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def sqroot(x):
    return np.sqrt(x)

x = np.linspace(0, 9)
y = sqroot(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def func3(x):
    return 3 * x ** 2 + 2 * x + 1

x = np.linspace(-2, 2)
y = func3(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def func4(x):
    return 4 * x ** 3 + 2 * x ** 2 + x + 3

x = np.linspace(-2, 2)
y = func4(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def f_sin(x):
    return np.sin(x)

def f_cos(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi)
y_sin = f_sin(x)
y_cos = f_cos(x)

plt.plot(x, y_sin, label = "sin")
plt.plot(x, y_cos, label = "cos")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def f_tan(x):
    return np.tan(x)

x = np.linspace(-1.3, 1.3)
y_tan = f_tan(x)

plt.plot(x, y_tan, label = "tan")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

def f_sin(x):
    return np.sin(x)

def f_cos(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi)
y_sin = np.abs(f_sin(x))
y_cos = np.abs(f_cos(x))

plt.plot(x, y_sin, label = "sin")
plt.plot(x, y_cos, label = "cos")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
