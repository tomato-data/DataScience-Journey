```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'JunYeol', 'Hanall', 'Junyoung'],
    'Age': [25, 30, 35, 27, 27, 27],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Male', 'Male'],
    'Height': [177, 178, 190, 176, 180, 180],
    'Weight': [80, 79, 10, 100, 20, 30]
}

df = pd.DataFrame(data)
df
```

```python
df_pivot = df.pivot(index = 'Gender', columns = 'Name', values = ['Height', 'Weight'])
df_pivot
```

```python
df_pivot.columns = ['Height_Alice', 'Height_Bob', 'Height_Charlie', 'Height_Hanall', 'Height_JunYeol', \
                    'Height_Junyoung', 'Weight_Ailice', 'Weight_Bob', 'Weight_Charlie', 'Weight_Hanall', \
                   'Weight_JunYeol', 'Weight_Junyoung']
df_pivot
```

```python
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago'],
    'Temperature': [70, 75, 65, 72, 68],
    'Humidity': [60, 55, 70, 62, 65]
}

df_city = pd.DataFrame(data)
grouped = df_city.groupby('City')
mean_values = grouped.mean()
max_values = grouped.max()
```
